```python
import re
def convert_to_list(x):
    return json.loads(x.replace('\'','\"')) if pd.notna(x) else x


def convert_to_list(x):
    if pd.notna(x):
        string = x.replace('\'','\"')
        corrected_text = re.sub(r'(?<=[a-zA-Z])\"(?=[a-zA-Z])', '\'', string) # to fix names like O"Brien after replace ' with "
        out = json.loads(corrected_text) 
    else:
        out = x
    return out
```
